```python
from rdkit import Chem
from hydra import initialize, compose
from pathlib import Path
from ergochem.draw import draw_molecule
from IPython.display import SVG
```

```python
with initialize(version_base=None, config_path="./conf/filepaths"):
    filepaths = compose(config_name="filepaths")
```

```python
chebi_gen = (mol for mol in Chem.SDMolSupplier(Path(filepaths.raw_data) / "ChEBI_lite_3star.sdf") if mol is not None)
```

```python
mol = next(
    chebi_gen
)  # Skip the first molecule, which is a blank line in the SDF file
```

```python
display(
    SVG(
        draw_molecule(mol)
    )
)
```
